```python
# to upload model 
import pickle

with open("neural_network_model.pkl", "rb") as file:
    nn_model = pickle.load(file)
```

```python
nn_model.summary()
```

```python
import cv2 
from ultralytics import YOLO
import numpy as np


# cap = cv2.VideoCapture('/Users/maxkucher/pytorch/YOLO/road.mp4')
# cap = cv2.VideoCapture('/Users/maxkucher/pytorch/YOLO/airport.mp4')
cap = cv2.VideoCapture('/Users/maxkucher/pytorch/YOLO/ped_zone.mp4')
model = YOLO("/Users/maxkucher/pytorch/yolov8n.pt")

names = model.names
threshold = 0.5


while True:

    car = 0
    bicycle = 0
    motorcycle = 0
    airplane = 0 
    bus = 0
    traffic_light = 0
    
    avaliable_objects = ["car", "bicycle", "motorcycle", "airplane", "bus", "traffic-light"]
    ret, frame = cap.read()
    results = model(frame)[0]
    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result
        if score > threshold and names[int(class_id)] in avaliable_objects:
            x_center = int((x1 + x2) / 2)
            y_center = int((y1 + y2) / 2)

            # draw a box around detected object
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            # draw a dot in the middle of detected object
            cv2.circle(frame, (x_center, y_center), 5, (0, 255, 0), thickness=cv2.FILLED)
            # put text under the box
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)


            if names[int(class_id)] == "car":
                car = 1
            else: 
                car = 0

            if names[int(class_id)] == "bicycle":
                bicycle = 1
            else:
                bicycle = 0


            if names[int(class_id)] == "motorcycle":
                motorcycle = 1
            else:
                motorcycle = 0


            if names[int(class_id)] == "airplane":
                airplane = 1
            else:
                airplane = 0

            if names[int(class_id)] == "bus":
                bus = 1
            else:
                bus = 0
            
            if names[int(class_id)] == "traffic-light":
                traffic_light = 1
            else:
                traffic_light = 0
            
            pred = nn_model.predict([[car, bicycle, motorcycle, airplane, bus, traffic_light]])
            zones = ["Road", "Airport", "Pedastrian zone"]
            
            cv2.putText(frame, f"Detected zone: {zones[np.argmax(pred)]}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

            

    


    if not ret:
        break
    
    # даем имя рамке и выводим ее 
    cv2.imshow("Object Detection", frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```
