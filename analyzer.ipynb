```python
x = data.drop("place", axis="columns")
y = data["place"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train = np.asarray(x_train).astype(np.float32)
y_train = np.asarray(y_train).astype(np.float32)


x_test = np.asarray(x_test).astype(np.float32)
y_test = np.asarray(y_test).astype(np.float32)
```

```python
nn_model = Sequential([
    Dense(9, input_shape=(6, ), activation = "relu"),
    Dense(8, activation = "relu"),
    Dense(5, activation = "relu"),
    Dense(4, activation = "relu"),
    Dense(3, activation = "softmax")


])

nn_model.compile(
    optimizer="SGD",
    loss = "sparse_categorical_crossentropy",
    metrics=(["accuracy"])
)


nn_model.fit(x_train, y_train, epochs=220)
```

```python
# IMPORTANT (about losses)
import matplotlib.pyplot as plt
history = nn_model.fit(x_train, y_train, epochs=250, verbose=0)

loss = history.history['loss']
epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'b', label='Loss')
plt.title("Loss/Epochs corelation")
plt.xlabel('Epochs')
plt.ylabel('Losses')
plt.legend()
plt.show()
```

```python
history = nn_model.fit(x_train, y_train, epochs=250, verbose=0)

accuracy = history.history['accuracy']
epochs = range(1, len(loss) + 1)

plt.plot(epochs, accuracy, 'b', label='Loss')
plt.title("Accuracy/Epochs corelation")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
nn_model.evaluate(x_test, y_test)
```
